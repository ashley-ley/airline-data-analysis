```python
import pandas as pd
from pathlib import Path
```

```python
us_weather_path = Path("united_states_weather.csv")
```

```python
us_weather_df = pd.read_csv(us_weather_path, encoding="ISO-8859-1")
```

```python
us_weather_df.head()
```

```python
us_weather_df[us_weather_df['City'] == 'Dallas'].count()
```

```python
dallas_weather = us_weather_df[us_weather_df['City'] == 'Dallas']
dallas_weather.head()
```

```python
us_weather_df = dallas_weather[['EventId','Type','Severity','StartTime(UTC)','AirportCode','LocationLat','LocationLng','City','County','State']]
us_weather_df
```

```python
# Assuming 'StartTime(UTC)' column is in datetime format
us_weather_df['StartTime(UTC)'] = pd.to_datetime(us_weather_df['StartTime(UTC)'])
```

```python
dallas_weather_df_2022 = us_weather_df[us_weather_df['StartTime(UTC)'].dt.year == 2022]
dallas_weather_df_2022.head()
```
