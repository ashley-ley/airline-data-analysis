```python
# Define the table_name parameter
params = {'table_name': 'flights'}
```

```python
# Send a GET request with the parameters
response = requests.get(base_url, params=params)
```

```python
import csv
import json
```

```python
with open('Weather_Data/five_country_flights.csv', mode='r', encoding='UTF-8') as csv_file:

    csv_reader = csv.DictReader(csv_file)
    

    json_data = []
    

    for row in csv_reader:
  
        json_data.append(row)
```

```python
with open('data.json', mode='w') as json_file:

    json.dump(json_data, json_file, indent=4)

print('CSV to JSON conversion completed.')
```

```python
if response.status_code == 200:
    try:
        data = response.json()
        flight_statuses = data.get('data.json', [])
        print(f'Flight Statuses: {flight_statuses}')
    except json.JSONDecodeError as e:
        print(f'Error decoding JSON: {e}')
else:
    print(f'Error: {response.status_code}, {response.text}')
```
